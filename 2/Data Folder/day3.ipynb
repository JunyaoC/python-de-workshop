```python
class Student():
    def __init__(self,name,age,gender):
        self.name = name
        self.age = age
        self.gender =gender
    
    def stu_name(self):
        return self.name

    def stu_age(self):
        return self.age

    def stu_gender(self):
        return self.gender
```

```python
student_4896 = Student("123",10,"Male")

print(student_4896.stu_name())
print(student_4896.stu_age())
print(student_4896.stu_gender())
```
