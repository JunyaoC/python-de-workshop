# Replace in Pandas

```python
replace_example = pd.DataFrame({'A': ['a','b','c'], 'B': [1.0,2.0,3.0]})
#replace_example

replace_example1 = replace_example.replace('b','e')
replace_example1
```

```python
replace_example2 = replace_example.replace({'A': 'b'}, 4.0)
replace_example2
```

```python
replace_example3 = replace_example.replace({'a':1.0 , 2.0 : 3.0 , 3.0:4.0})
replace_example3
```
