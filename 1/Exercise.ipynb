# Data Engineering Exercie Part 1

1. Import the dataset "vaccinations.csv"

```python
import pandas as pd
```

```python
def check_df(df):
    display(df.head())
    print(F"Columns: {len(df.columns)}")
```

```python
vaccinationDF = pd.read_csv("vaccinations.csv")
```

```python
check_df(vaccinationDF)
```

2. Drop "daily_vaccinations_raw"column

```python
vaccinationDF = vaccinationDF.drop(columns = ["daily_vaccinations_raw"])
check_df(vaccinationDF)
```

3. Rename the column "Location" to "Country"

```python
vaccinationDF = vaccinationDF.rename(columns = {"location":"country"})
check_df(vaccinationDF)
```

```python
censored = ["World","Asia","Upper middle income"]

for x in censored:
    vaccinationDF = vaccinationDF.query(F"country != {censored}")
```

4. Find the country that has the highest count of total_vaccinations

```python
vaccinationDF[ vaccinationDF["total_vaccinations"] == vaccinationDF.total_vaccinations.max() ]
```

```python
vaccinationDF.query(F" total_vaccinations == {vaccinationDF.total_vaccinations.max()} ")
```

5. List the data that has the count of total_vaccinations less than 10,000 which country has the lowest among the data?

```python
lt10kDF = vaccinationDF[ vaccinationDF["total_vaccinations"] < 10000 ].sort_values(by = ["total_vaccinations"], ascending = True)
lt10kDF[["country","total_vaccinations"]]
```

```python
lt10kDF = vaccinationDF.query(" total_vaccinations < 10000 ").sort_values(by = ["total_vaccinations"], ascending = True)
lt10kDF[["country","total_vaccinations"]]
```

```python
lt10kDF[ lt10kDF["total_vaccinations"] == lt10kDF.total_vaccinations.min() ][["country","total_vaccinations"]]
```
